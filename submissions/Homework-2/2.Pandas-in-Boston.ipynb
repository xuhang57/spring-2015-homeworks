## Step 3. Advanced filtering & joins

By now, you should have noticed that, for each day, there are multiple records in our dataframe. Use the `groupby` function to get, for each day, the counts for all the case categories. Plot the result for the top 5 categories as you see fit. **(20 pts)**

````python
# Write your code here
# days = boston.groupby(pd.Grouper(key="open_dt", freq="D"))
# daily_titles = {}
# for day, daily_data in days:
#     daily_titles[day] = pd.value_counts(daily_data["case_title"])
      
#       # Count of each title per a given day, indexed by day (days x titles)
# daily_titles = pd.DataFrame.from_dict(daily_titles, orient="index")

#       # Sum each column across all rows and get top 5 titles
# top_titles = daily_titles.sum(axis=0).order(ascending=False).head(5)
      
# top_titles.plot()

#Grabs top 5 occuring case title categories
top_5 = boston['case_title'].value_counts().head(5).index
     
#Filters out CASE_TITLES not in top_5
top5_only = boston[boston.case_title.isin(top_5)]
  
#Groups by OPEN_DT
top5_only.groupby(boston.open_dt.dt.date)
top5_only.index = top5_only.open_dt.dt.date
 
gb_date = boston.groupby(boston.open_dt.dt.date)
requests = {}

for date, case in gb_date:
    requests[date] = {}
    for title in boston['case_title'].value_counts().head(5).index:
        date_title = case.case_title
        requests[date][title] = date_title[date_title == title].count()

requests = sorted(requests.items())
   
index = sorted(top5_only.open_dt.dt.date.value_counts().index)
cols = boston['case_title'].value_counts().head(5).index
    
sorted_frame = pd.DataFrame(index=index, columns = cols)
      
for date, cases in requests:
    for val in cases:
        sorted_frame.ix[date][val] = cases[val]
print sorted_frame
sorted_frame.plot(figsize=(12,5))
        

        
````

Let's see how this trend compares to the snowfall. First, we need to read the data from `snow.csv`

````python
# Read the dataframe
snow_data = pd.read_csv("snow.csv",parse_dates=[1])
print snow_data
````

According to the documentation, the column `SNOW` has the values for the snowfall, in mm. However, the dataset we have has information from many different stations around Boston. Aggregate all their measurements and transform the result to centimeters. Finally, plot the result **(10 pts)**

````python
# # Write your code here
# daily_snow_data = snow_data.groupby("DATE")

# # Daily snow measurements, indexed by day
# daily_snow_measurements = {}
# for day, daily_data in daily_snow_data:
#     # Aggregate each station's snowfall per day and convert from mm to cm
#     daily_snow_measurements[day.date()] = daily_data[daily_data["SNOW"] >= 0]["SNOW"].sum() / 10
    
    
      
#       # Convert from dict to DataFrame
# daily_snow_measurements = pd.DataFrame.from_dict(daily_snow_measurements, orient="index").sort_index()
# daily_snow_measurements.columns = ["SNOWFALL"]
      
# daily_snow_measurements.plot(kind = 'barh')

#Converts to cm\n",
snow_data.SNOW = snow_data.SNOW.apply(lambda c: c/10.0)
#Grabs dates with only positive snowfall
only_pos = snow_data[snow_data.SNOW >= 0]

      #Groups by date
gb_date = only_pos.groupby(only_pos.DATE.dt.date)

snowfall = {}

      #Creates a dictionary with the date as the keys and mean snowfall as the value
for date,snow in gb_date:
    snowfall[date] = snow.SNOW.mean()
      
#Gets the keys (dates) as the index for the new dataframe
index = sorted(snowfall.keys())
      
#Creates a data frame of the mean snowfall
mean_sf = pd.DataFrame.from_dict(sorted(snowfall.items()))
        
#Makes the index of the new dataframe the keys(dates)
mean_sf.index = index
mean_sf = mean_sf.drop([0],axis=1)

#Renames the column
mean_sf.columns = ['Average Snowfall']

#Plots the mean snowfall over time
mean_sf.plot(figsize=(12,5),label = 'average snowfall')
# print snow_data
````

````python
# "daily_top_titles = pd.DataFrame(index=daily_snow_measurements.index, columns=top_titles.index)\n",
#       "\n",
#       "for day, row in daily_top_titles.iterrows():\n",
#       "    for title in daily_top_titles:\n",
#       "        daily_top_titles.ix[day][title] = daily_titles.ix[day][title]\n",
#       "    \n",
#       "daily_top_titles = daily_top_titles.fillna(value=0)\n",
#       "\n",
#       "calls_on_snow = daily_top_titles.join(daily_snow_measurements)\n",
#       "\n",
#       "calls_on_snow"
        
        
#Combines the dataframes
calls_on_snow = pd.concat([mean_sf,sorted_frame],axis=1)
     
#Fills all NaN with 0
calls_on_snow.fillna(0,inplace=True)
      
#Sorts the columns so that average snowfall is 6th column
calls_on_snow = calls_on_snow[['Misc. Snow Complaint','Request for Pothole Repair','Request for Snow Plowing','Requests for Street Cleaning','Snow Removal','Average Snowfall']]

calls_on_snow
````
