```python
from nbastats.nbastats import ShotChart, PlayByPlay, BoxScore
```

```python
gameid = '0021300001' #Magic vs. Pacers 10-29-2013
```

```python
box = BoxScore(gameid)
box.gameinfo()
```

```python
active_players = box.playerstats()[['PLAYER_ID','PLAYER_NAME','TEAM_ABBREVIATION']].ix[(box.playerstats()['MIN']>=0)]
active_ids = active_players.PLAYER_ID.tolist() #Need to create a list of just ID numbers to iterate over
active_players
```

```python
plays = PlayByPlay(gameid).pbp()
```

```python
import pandas as pd
shots = pd.DataFrame()
for x in active_ids:
    try:
        shots = shots.append(ShotChart(x,gameid=gameid).shotchart())
    except:
        pass
```
